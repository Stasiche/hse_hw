```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_subsample = 100
size_subsample = 1000

k_max = 100
```

```python
# uniform distribution
rmse_lst = []
us = np.random.uniform(size=(num_subsample, size_subsample))
for k in range(1, k_max):
    theta_star = np.power((k+1)*np.mean(np.power(us,k), axis=1), 1/k)
    rmse = np.sqrt(np.mean(np.power(np.ones(len(theta_star))-theta_star,2)))
    rmse_lst.append(rmse)
plt.plot(range(1,k_max), rmse_lst);
```

```python
# exponentinal distribution
rmse_lst = []
es = np.random.exponential(size=(num_subsample, size_subsample))
for k in range(1, k_max):
    theta_star = np.power(np.mean(np.power(es,k)/np.math.factorial(k), axis=1), 1/k)
    rmse = np.sqrt(np.mean(np.power(np.ones(len(theta_star))-theta_star,2)))
    rmse_lst.append(rmse)
plt.plot(range(1,k_max), rmse_lst);
```

## Вывод
Как видно из графиков выше, для оценки масшатбного параметра равномерного распределения лучше брать моменты более высокого порядка, а для экспоненциального более низкого.
